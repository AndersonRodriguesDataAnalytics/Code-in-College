```python
#Aplicando uma classe
class Pessoa:
    def __init__(self, Nome, Idade):
      self.Nome = Nome
      self.Idade = Idade

    def BoasVindas(self):
      print("Seja bem vindo: " + self.Nome)

    def Recusado(self):
      print("Acesso negado")


    #função

    def Maior(self):
      if self.Idade >= 18:
        print("Você é maior de idade")
        self.BoasVindas()
      else:
        print("Você é menor de idade")
        self.Recusado()
```

```python
Dados = Pessoa('Zeca', 19)
```

```python
Dados = Dados.Maior()
```
